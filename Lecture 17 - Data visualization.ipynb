# Data Visualization

* Pandas is not a data visualization library but it is capable of creating basic plots. 
* If you are just creating plots for exploratory data analysis, Pandas might be highly useful and practical.

* Dataset is available on Kaggle with a creative commons license.

https://www.kaggle.com/datasets/mathchi/churn-for-bank-customers

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd

churn = pd.read_csv("Data/churn.csv")

churn.head()
```

```python
churn.shape
```

## Exercise 1 - take a sample from the dataset

```python
churn = churn[churn["Balance"]>0].sample(n=500, ignore_index=True)

churn.head()
```

```python
churn.shape
```

## Exercise 2 - scatter plot

```python
churn.plot(x="Balance", y="EstimatedSalary", kind="scatter")
```

## Exercise 3 - change the size

```python
churn.plot(
    
    x="Balance", 
    y="EstimatedSalary", 
    kind="scatter",
    figsize=(7, 4)
    
)
```

## Exercise 4 - add a title

```python
churn.plot(
    
    x="Balance", 
    y="EstimatedSalary", 
    kind="scatter",
    figsize=(7, 4),
    title="Estimated Salary vs Balance"
    
)
```

## Exercise 5 - axis labels

```python
churn.plot(
    
    x="Balance", 
    y="EstimatedSalary", 
    kind="scatter",
    figsize=(7, 4),
    title="Estimated Salary vs Balance",
    ylabel="Salary",
    xlabel="Monthly Balance"
    
)
```

## Exercise 6 - Histogram

```python
churn["Balance"].plot(
    
    kind="hist",
    figsize=(7, 5),
    title="Distribution of Balances"

)
```

## Exercise 7 - hist function

```python
churn.hist(
    
    "Balance",
    figsize=(8, 4)

)
```

## Exercise 8 - by parameter

```python
churn.head()
```

```python
churn.hist(
    
    "Balance",
    figsize=(10, 4),
    by="Gender"

)
```

## Exercise 9 - sharey parameter


```python
churn.hist(
    
    "Balance",
    figsize=(10, 4),
    by="Gender",
    sharey=True

)
```

## Exercise 10 - more groups

```python
churn.head()
```

```python
churn.hist(
    
    "Balance",
    figsize=(7, 4),
    by="Geography",
    sharey=True

)
```

## Exercise 11 - layout parameter

```python
churn.hist(
    
    "Balance",
    figsize=(10, 3),
    by="Geography",
    sharey=True,
    layout=(1, 3)

)
```

## Exercise 12 - labelsize and rotation

```python
churn.hist(
    
    "Balance",
    figsize=(10, 3),
    by="Geography",
    sharey=True,
    bins=8,
    layout=(1, 3),
    xlabelsize=10,
    ylabelsize=10,
    xrot=45

)
```

## Exercise 13 - stock price dataset

```python
df = pd.read_csv("Data/apple_google_stock_prices_092022.csv")

df.head()
```

```python
df = df[["Date", "Close", "Stock"]]

df.head()
```

```python
df_pivoted = df.pivot_table(index="Date", columns="Stock")

df_pivoted.head()
```

```python
df_pivoted.columns
```

## Exercise 14 - line plot

```python
df_pivoted.plot.line()
```

```python
df_pivoted.plot.line(title="Apple and Google Stock Prices", ylabel="USD")
```
